```python
# Create a date range with both date and time
start_date = '1996-01-01 00:00:00'
end_date = '2023-05-18 23:59:59'  # Specify the end time as the end of the day
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Add the timestamp column to your DataFrame
mp['Timestamp'] = date_range

# Display the updated DataFrame
mp
```

```python
# Import Libraries for LSTM Model
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
import statsmodels.api as sm

# Choose a specific column as your time series data
date_column = 'Timestamp'

# Convert the date column to datetime
mp[date_column] = pd.to_datetime(mp[date_column])

# Create a numerical time index based on the date (e.g., using day numbers)
mp['TimeIndex'] = (mp[date_column] - mp[date_column].min()).dt.days

# Specify the period based on your data (e.g., 365 for daily data)
period = 365

# Perform classical time series decomposition with the specified period
decomposition = sm.tsa.seasonal_decompose(mp[date_column], model='additive', period=period)

# Extract the trend, seasonal, and residual components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(mp[date_column], label='Original', color='blue')
plt.legend(loc='upper left')
plt.title('Original Time Series')

plt.subplot(412)
plt.plot(mp[date_column], trend, label='Trend', color='red')
plt.legend(loc='upper left')
plt.title('Trend Component')

plt.subplot(413)
plt.plot(mp[date_column], seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.title('Seasonal Component')

plt.subplot(414)
plt.plot(mp[date_column], residual, label='Residual', color='purple')
plt.legend(loc='upper left')
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
# Function to check stationarity using Augmented Dickey-Fuller test
def check_stationarity(ts):
    result = adfuller(ts)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'{key}: {value}')

# Check stationarity of the time series
check_stationarity(mp['Timestamp'])

# If not stationary, apply differencing
mp['DifferencedColumn'] = mp['Tool wear [min]'].diff().fillna(0)  # Use 'YourTimeSeriesColumn'
check_stationarity(mp['DifferencedColumn'])
```

In summary, the ADF test results suggest that  time series is stationary. This is important because many time series models,
including ARIMA and LSTM, assume stationarity for accurate forecasting. We can proceed with modeling and forecasting using this
stationary time series.

```python
# Plot ACF and PACF for the 'Tool wear [min]' column after differencing
plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(mp['Tool wear [min]'], lags=40, ax=plt.gca(), title='ACF')
plt.subplot(212)
plot_pacf(mp['Tool wear [min]'], lags=40, ax=plt.gca(), title='PACF')
plt.tight_layout()

plt.show()
```

We can start with an ARIMA(p, d, q) model with p = 2, d = (0),
and q = 17.

```python
columns_to_convert = ['Type']
# Loop through the list of columns and convert them
for column in columns_to_convert:
    mp[column] = pd.Categorical(mp[column]).codes
mp
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA  # Import ARIMA from the new location
from statsmodels.tsa.statespace.sarimax import SARIMAX  # You can also use SARIMAX for more flexibility
# 1. Select Predictor Variables
predictor_cols = ['Type', 'OSF', 'PWF', 'HDF', 'TWF', 'Torque [Nm]', 'Tool wear [min]']

# 2. Prepare Data
# Create a new DataFrame with both the time series and predictor variables
data = mp[['Machine failure'] + predictor_cols]

# 3. Train-Validation-Test Split
train_ratio = 0.7
validation_ratio = 0.15
test_ratio = 0.15

train_size = int(len(data) * train_ratio)
validation_size = int(len(data) * validation_ratio)

train_data = data[:train_size]
validation_data = data[train_size:train_size + validation_size]
test_data = data[train_size + validation_size:]

# 4. Fit ARIMA Model with Exogenous Variables
p, d, q = 2, 0, 17  # Define ARIMA order
exog_train = train_data[predictor_cols]  # Exogenous variables for training

# Fit ARIMA model with exogenous variables
model = ARIMA(train_data['Machine failure'], order=(p, d, q), exog=exog_train)
model_fit = model.fit()

print(model_fit.summary())
```

CHECKING RESIDUALS

```python
residuals = model_fit.resid
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(residuals, lags=40)  # Adjust the number of lags as needed
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(residuals, lags=40)  # Adjust the number of lags as needed
```

PREDICTION

```python
# Define the number of periods (time steps) into the future you want to predict
n_periods = 11  # Adjust this value according to your prediction needs

# Prepare exogenous variables for prediction
# Ensure predictor_data has the same format as used for training
predictor_data = mp[['Type', 'OSF', 'PWF', 'HDF', 'TWF', 'Torque [Nm]', 'Tool wear [min]']]
predictor_data = predictor_data[-n_periods:]  # Select the last n_periods rows for forecasting

# Generate forecasts for the specified number of periods
start = len(model_fit.fittedvalues) + 1
end = start + n_periods - 1
forecast = model_fit.forecast(steps=n_periods, exog=predictor_data)

# Print the forecasts
print("Predicted Machine Failure:")
print(forecast)
```
